```python
def shortner(category, cutoff):
    cat_map = {}
    for i in range(len(category)):
        if category.values[i] >= cutoff:
            cat_map[category.values[i]] = category.index[i]
    return cat_map

shortner(df.groupby('Country_name')['Price_Paid'].sum().sort_values(),146240608).values
```
